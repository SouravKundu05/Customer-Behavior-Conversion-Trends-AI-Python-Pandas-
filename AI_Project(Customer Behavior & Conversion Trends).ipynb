```python
#How does the frequency of events like product_view, add_to_cart, and buy_button_click vary across customers with statuses New, Active, and Inactive? 

# Prompt :- is "Analyze the frequency of each event_name for different customer statuses (New, Active, Inactive). 
#           The customer_status column is in the customers_df, while event_name is in the customer_activity_df. 
#           Use groupings and pivot tables to present results clearly, with event_name as rows and customer_status as columns."

result = (
    customers_activity_df
    .merge(
        customers_df[['customer_id','customer_status']],
        on='customer_id',
        how='left'
    )
    .groupby(['event_name','customer_status'])
    .size()
    .unstack(fill_value=0)
)

result
```

```python
# How does the total number of activities logged by a customer correlate with their total number of purchases? 

#Prompt :- "Count the total number of activities (event_name) each customer has performed and analyze its relationship with total_purchases 
#           from the Customers Table. Calculate the correlation between total activities and purchases to understand if engagement drives purchases." 

analysis_df = (
    customers_activity_df
    .groupby('customer_id')
    .size()
    .reset_index(name='total_activities')
    .merge(
        customers_df[['customer_id','total_purchases']],
        on='customer_id',
        how='left'
    )
)

correlation = analysis_df['total_activities'].corr(
    analysis_df['total_purchases']
)

print(correlation)
```

```python
# What is the average time taken by customers to make their first buy_button_click after signing up? 
# How does this vary across locations and referral sources? 

# Prompt :- "For customers who performed the buy_button_click event, calculate the number of days between their signup_date (from customers_df) 
#          and their first buy_button_click date (from customer_activity_df). Group the results by location to find average purchase timelines for 
#          each location." 

result = (
    customers_activity_df
    .query("event_name == 'buy_button_click'")
    .groupby('customer_id')['event_datetime']
    .min()
    .reset_index(name='first_buy_date')
    .merge(
        customers_df[['customer_id','signup_date','location']],
        on='customer_id',
        how='left'
    )
    .assign(
        days_to_first_buy=lambda x:
        (x['first_buy_date'] - x['signup_date']).dt.days
    )
    .groupby('location')['days_to_first_buy']
    .mean()
    .reset_index(name='avg_days_to_first_buy')
    .sort_values('avg_days_to_first_buy')
)

result
```

```python
# Analyze the number of product_view and add_to_cart events across salary ranges (e.g., <30K, 30K-60K, 60K-100K, >100K). 

# Prompt :- "Group customers into salary ranges (e.g., <30K, 30K-60K, 60K-100K, >100K) using the salary column. Analyze how the 
#            frequency of events (event_name) varies across these salary ranges. Present the results in a pivot table format, showing 
#            event names as rows and salary ranges as columns." 

result = (
    customers_activity_df
    .merge(
        customers_df.assign(
            salary_range=pd.cut(
                customers_df['salary'],
                bins=[0,30000,60000,100000,float('inf')],
                labels=['<30K','30K-60K','60K-100K','>100K']
            )
        )[['customer_id','salary_range']],
        on='customer_id',
        how='left'
    )
    .groupby(['event_name','salary_range'])
    .size()
    .unstack(fill_value=0)
)

result
```

```python
# Which preferred device (Mobile, Tablet, Desktop) leads to the highest percentage of buy_button_click events? 

# Prompt :- "Determine the conversion rate for each preferred_device (Mobile, Tablet, Desktop). Conversion rate is defined as 
#            the number of buy_button_click events divided by the total number of events for each device." 

result = (
    customers_activity_df
    .merge(
        customers_df[['customer_id','preferred_device']],
        on='customer_id',
        how='left'
    )
    .assign(
        is_buy=lambda x: x['event_name'] == 'buy_button_click'
    )
    .groupby('preferred_device')
    .agg(
        total_events=('event_name','count'),
        buy_events=('is_buy','sum')
    )
    .assign(
        conversion_rate=lambda x:
        x['buy_events'] / x['total_events']
    )
    .reset_index()
)

result
```

```python
# How do event activities (product_view, wishlist, add_to_cart) differ across customer locations (Metro, Tier 1, Tier 2, Rural)? 

# Prompt :- "Analyze the distribution of events (event_name) across customer location categories (Metro, Tier 1, etc.). 
#            Provide the total counts of each event type for each location in a pivot table format." 

result = (
    customers_activity_df
    .merge(
        customers_df[['customer_id','location']],
        on='customer_id',
        how='left'
    )
    .groupby(['event_name','location'])
    .size()
    .unstack(fill_value=0)
)

result
```

```python
# Do customers who spend more time on activities have higher total spending? Analyze the relationship between average event_duration and total_spend. 

# Prompt :- "Calculate the average event_duration for each customer and compare it to their total_spend. Analyze the correlation between 
#            session duration and spending to determine if engagement time impacts spending." 

analysis_df = (
    customers_activity_df
    .groupby('customer_id')['event_duration']
    .mean()
    .reset_index(name='avg_event_duration')
    .merge(
        customers_df[['customer_id','total_spend']],
        on='customer_id',
        how='left'
    )
)

correlation = analysis_df['avg_event_duration'].corr(
    analysis_df['total_spend']
)

print(correlation)
```

```python
# What events are most common among the top 10% of customers by total spending? 

# Prompt :- "Analyze the conversion funnel by counting the number of occurrences of events in the order 
#            product_view → add_to_cart → buy_button_click. Identify where most users drop off in the funnel." 

funnel_summary = (
    customers_activity_df
    .query("event_name in ['product_view','add_to_cart','buy_button_click']")
    .drop_duplicates(['customer_id','event_name'])
    .pivot(index='customer_id', columns='event_name', values='event_datetime')
    .notna()
    .sum()
    .rename_axis('stage')
    .reset_index(name='users')
)

funnel_summary['drop_off_%'] = (
    funnel_summary['users'].pct_change() * -100
)

funnel_summary
```

```python
# Among customers who performed product_view, how many proceeded to add_to_cart,and how many completed a buy_button_click? Identify drop-off points. 

# Prompt :- "Analyze whether customers who scrolled (is_scroll = True) during activities are more likely to 
#            click the buy_button_click event. Compare conversion rates for scrollers vs. non-scrollers." 

analysis_df = (
    customers_activity_df
    .assign(
        is_buy=lambda x: x['event_name']=='buy_button_click'
    )
    .groupby('customer_id')
    .agg(
        is_scroller=('is_scroll','max'),
        did_buy_click=('is_buy','max')
    )
    .reset_index()
)

result = (
    analysis_df
    .groupby('is_scroller')
    .agg(
        total_customers=('customer_id','count'),
        buyers=('did_buy_click','sum')
    )
    .assign(
        conversion_rate=lambda x:
        x['buyers']/x['total_customers']
    )
    .reset_index()
)

result
```

```python
# Are customers who scroll during activities more likely to reach buy_button_click? Compare the proportion of scroll events with conversions. 

# Prompt :- "Identify customers who have not been active in the last 30 days by calculating the difference between today’s date and 
#            their most recent activity (event_datetime). Provide a list of such churned customers along with the days since their last activity." 

today = pd.Timestamp.today().normalize()

churned_customers = (
    customers_activity_df
    .groupby('customer_id')['event_datetime']
    .max()
    .reset_index(name='last_activity_date')
    .assign(
        days_since_last_activity=lambda x:
        (today - x['last_activity_date']).dt.days
    )
    .query("days_since_last_activity > 30")
    .merge(
        customers_df[['customer_id','name','location','customer_status']],
        on='customer_id',
        how='left'
    )
)

churned_customers
```

```python
# How many customers have no activity recorded in the last 30 days? What are their demographics (e.g., age, location, referral source)? 

# Prompt :- "Find the total number of activities for each customer location and sort the results in descending order to identify the most 
#            active locations." 

result = (
    customers_activity_df
    .merge(
        customers_df[['customer_id','location']],
        on='customer_id',
        how='left'
    )
    .groupby('location')
    .size()
    .reset_index(name='total_activities')
    .sort_values('total_activities', ascending=False)
)

result
```

```python
# Which referral sources (Organic, Ad Campaign, etc.) lead to the highest percentage of customers performing buy_button_click? 

# Prompt :- "Calculate the conversion rate for each referral_source by dividing the number of customers who clicked buy_button_click by the 
#            total number of customers from that referral source." 

result = (
    customers_df[['customer_id','referral_source']]
    .merge(
        customers_activity_df
        .assign(is_buy=lambda x: x['event_name']=='buy_button_click')
        .groupby('customer_id')['is_buy']
        .max()
        .reset_index(name='did_buy_click'),
        on='customer_id',
        how='left'
    )
    .assign(did_buy_click=lambda x: x['did_buy_click'].fillna(False))
    .groupby('referral_source')
    .agg(
        total_customers=('customer_id','count'),
        buyers=('did_buy_click','sum')
    )
    .assign(
        conversion_rate=lambda x:
        x['buyers']/x['total_customers']
    )
    .reset_index()
    .sort_values('conversion_rate', ascending=False)
)

result
```

```python
#For each signup month, what is the average number of activities per customer? Is there a trend over time? 

# Prompt :- "Analyze the total number of events (event_name) performed by customers grouped by their signup month. 
#            Show trends in user activity over time." 

result = (
    customers_activity_df
    .merge(
        customers_df.assign(
            signup_month=lambda x:
            x['signup_date'].dt.to_period('M')
        )[['customer_id','signup_month']],
        on='customer_id',
        how='left'
    )
    .groupby('signup_month')
    .size()
    .reset_index(name='total_events')
    .sort_values('signup_month')
)

result
```

```python
# Which pages (Home, Product Page, Cart Page, Checkout Page) see the most buy_button_click events?

# Prompt :- "Determine which page_name (e.g., Product Page, Checkout Page) has the highest conversion rate by analyzing the count of 
#            buy_button_click events for each page." 

result = (
    customers_activity_df
    .assign(
        is_buy=lambda x: x['event_name']=='buy_button_click'
    )
    .groupby('page_name')
    .agg(
        total_events=('event_name','count'),
        buy_events=('is_buy','sum')
    )
    .assign(
        conversion_rate=lambda x:
        x['buy_events']/x['total_events']
    )
    .reset_index()
    .sort_values('conversion_rate', ascending=False)
)

result
```

```python
# For customers with fewer than 10 activities, what percentage complete buy_button_click compared to customers with more than 20 activities? 

# Prompt :- "Categorize customers based on their total activity volume (Low, Medium, High, Very High) using activity counts (event_name). 
#            Provide the number of customers in each category." 

result = (
    customers_activity_df
    .groupby('customer_id')
    .size()
    .reset_index(name='total_activities')
    .assign(
        activity_category=lambda x: pd.cut(
            x['total_activities'],
            bins=[0,10,30,60,float('inf')],
            labels=['Low','Medium','High','Very High']
        )
    )
    .groupby('activity_category')
    .size()
    .reset_index(name='num_customers')
    .sort_values('num_customers', ascending=False)
)

result
```

```python
# Which devices are most commonly used for activities like product_view, wishlist, and buy_button_click? 

# Prompt :- "Analyze how events (event_name) are distributed across preferred_device categories (Mobile, Tablet, Desktop). 
#            Provide a total count of events for each device type." 

result = (
    customers_activity_df
    .merge(
        customers_df[['customer_id','preferred_device']],
        on='customer_id',
        how='left'
    )
    .groupby('preferred_device')
    .size()
    .reset_index(name='total_events')
    .sort_values('total_events', ascending=False)
)

result
```

```python
# How does the frequency of events (login, product_view, etc.) vary across app versions? 

# Prompt :- "Find the frequency of events (event_name) for each app_version. Summarize the results and include the total event 
#            count for each app version."

pivot_table = (
    customers_activity_df
    .groupby(['app_version','event_name'])
    .size()
    .unstack(fill_value=0)
)

pivot_table.loc['Total_Events'] = (
    customers_activity_df
    .groupby('app_version')
    .size()
)

pivot_table
```

```python
# Do customers with a higher number of add_to_cart events spend more on average than those with fewer such events? 

# Prompt :- "Group customers by their total number of activities into categories: Low, Medium, High, and Very High. 
#            Calculate the average total_spend for each activity category."

result = (
    customers_activity_df
    .groupby('customer_id')
    .size()
    .reset_index(name='total_activities')
    .assign(
        activity_category=lambda x: pd.cut(
            x['total_activities'],
            bins=[0,10,30,60,float('inf')],
            labels=['Low','Medium','High','Very High']
        )
    )
    .merge(
        customers_df[['customer_id','total_spend']],
        on='customer_id',
        how='left'
    )
    .groupby('activity_category')
    .agg(
        avg_total_spend=('total_spend','mean'),
        num_customers=('customer_id','count')
    )
    .reset_index()
    .sort_values('avg_total_spend', ascending=False)
)

result
```

```python
# Which locations contribute the most to total spending and have the highest number of buy_button_click events? 

# Prompt :- "Identify the top-performing locations based on the number of unique customers who completed the buy_button_click event." 

result = (
    customers_activity_df
    .query("event_name == 'buy_button_click'")
    .merge(
        customers_df[['customer_id','location']],
        on='customer_id',
        how='left'
    )
    .groupby('location')['customer_id']
    .nunique()
    .reset_index(name='unique_buyers')
    .sort_values('unique_buyers', ascending=False)
)

result
```

```python
# How do New and Active customers differ in the number of activities they perform and their likelihood of reaching buy_button_click? 

# Prompt :- "Analyze the behavior of customers by their customer_status (New, Active, Inactive). Provide a breakdown of event counts 
#            for each status group."

pivot_table = (
    customers_activity_df
    .merge(
        customers_df[['customer_id','customer_status']],
        on='customer_id',
        how='left'
    )
    .groupby(['event_name','customer_status'])
    .size()
    .unstack(fill_value=0)
)

pivot_table
```
